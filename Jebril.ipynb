```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data['alcohol'], bins=20)
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.title('Distribution of Alcohol Content')
plt.show()
```

## Histogram of Alcohol Content
The histogram shows that the alcohol content is normally distributed with a peak around 10-11% alcohol content.

```python
plt.scatter(data['alcohol'], data['quality'])
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.title('Scatter Plot of Alcohol Content vs. Quality')
plt.show()
```

## Scatter plot of Alcohol content vs the Quality
A correlation between the Alcohol content and the Quality as indicate in the scatter plot shows higher alcohol level is linked to higher quality evaluations.
